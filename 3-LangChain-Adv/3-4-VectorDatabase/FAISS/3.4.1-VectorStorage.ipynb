# cosine_similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = ["What is the capital of USA?", 
             "Who is a president of USA?", 
             "Who is a prime minister of India?"]
```

```python
# Check similarity between two sentences
my_query = "Narendra modi is prime minister of India"
```

```python
# To check similarity we first need to do embedding of this sentences.
documents_embedding = embeddings.embed_documents(documents)
```

```python
documents_embedding
```

```python
query_embedding = embeddings.embed_query(my_query)
```

```python
query_embedding
```

```python
cosine_similarity([query_embedding], documents_embedding)
```

```python
# In the above output we can clearly see that 3rd sentence has highest similarity with the query
```

# Euclidean distance

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
euclidean_distances([query_embedding], documents_embedding)
```

```python
# in euclidean distance the most nearest sentence is matched and it scores it lowest because the sentence is in lowest distance (so the 3rd sentence is again nearest one)
```

| Metric            | Similarity Score Range | Behavior                              |
| ----------------- | ---------------------- | ------------------------------------- |
| Cosine Similarity | \[-1, 1]               | Focuses on angle only |
| L2 Distance       | \[0, )                | Focuses on **magnitude + direction**  |


# Create Index

### Index is nothing but we can say that it is a way to ref a data

```python
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
# create an index
index = faiss.IndexFlatL2(384)

# How we decided 384 here?
# use len(query_embedding)

# This index is created using L2 Distance
len(query_embedding)
```

```python
index
```

```python
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)
```

```python
vector_store.add_texts(["AI is future","AI is powerful","Dogs are cute"])
```

```python
# So here above for each sentence we go one UUID (basically a unique id for each sentence)
```

```python
# How to get UUID of all sentence/documents?
vector_store.index_to_docstore_id
```

```python
results = vector_store.similarity_search("Tell me about AI", k=2)

# here arg1: is query and arg2: num of document we want in response
```

```python
results
```

```python
# Now here if ask for 3 documents then the lowest similar will be at last (a dog one)
results = vector_store.similarity_search("Tell me about AI", k=3)
results
```

# Type of Index
1. Flat Index
2. HNSW Index
3. IVF Index

1. Flat Index
- we navigate each and every record
- So this is called exact match

2. HNSW & IVF
- Approximate Match (ANN)
- Not navigating to each documents
- HNSW is a graph based mechanism
- IVF is a cluster based mechanism

| Feature               | `Flat`                | `IVF` (Inverted File Index)        | `HNSW` (Graph-based Index)          |
| --------------------- | --------------------- | ---------------------------------- | ----------------------------------- |
| Type of Search     | Exact                 | Approximate (cluster-based)        | Approximate (graph-based traversal) |
| Speed               | Slow (linear scan)    | Fast (search only in top clusters) | Very Fast (graph walk)              |


| Dataset Size              | Recommended Index                 |
| ------------------------- | --------------------------------- |
| UPTO 1L                     | `IndexFlatL2` or `IndexFlatIP`    |
| UPTO 1M                  | `IndexIVFFlat` or `IndexHNSWFlat` |
| > 1M                      | `IndexIVFPQ` or `IndexHNSWFlat`   |


```python
# from uuid import uuid4
from langchain_core.documents import Document

document_1 = Document(
    page_content="I had chocolate chip pancakes and scrambled eggs for breakfast this morning.",
    metadata={"source": "tweet"},
)

document_2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
)

document_3 = Document(
    page_content="Building an exciting new project with LangChain - come check it out!",
    metadata={"source": "tweet"},
)

document_4 = Document(
    page_content="Robbers broke into the city bank and stole $1 million in cash.",
    metadata={"source": "news"},
)

document_5 = Document(
    page_content="Wow! That was an amazing movie. I can't wait to see it again.",
    metadata={"source": "tweet"},
)

document_6 = Document(
    page_content="Is the new iPhone worth the price? Read this review to find out.",
    metadata={"source": "website"},
)

document_7 = Document(
    page_content="The top 10 soccer players in the world right now.",
    metadata={"source": "website"},
)

document_8 = Document(
    page_content="LangGraph is the best framework for building stateful, agentic applications!",
    metadata={"source": "tweet"},
)

document_9 = Document(
    page_content="The stock market is down 500 points today due to fears of a recession.",
    metadata={"source": "news"},
)

document_10 = Document(
    page_content="I have a bad feeling I am going to get deleted :(",
    metadata={"source": "tweet"},
)

documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]
```

```python
index = faiss.IndexFlatIP(384)

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(), 
    index_to_docstore_id={}
    )
```

```python
vector_store.add_documents(documents)
```

```python
vector_store.similarity_search(
    "Langchain provides abstractions to make working with LLMs easy", 
    k=2
    )
```

```python
vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
    filter={"source":{"$eq": "tweet"}}    
)
```

```python
result=vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
    filter={"source":"news"}    
)
```

```python
result[0].metadata
```

```python
result[0].page_content
```

```python
# We are creating this retriever so that we can use this further into our RAG pipeline
retriever=vector_store.as_retriever(search_kwargs={"k": 3})
```

```python
retriever.invoke("LangChain provides abstractions to make working with LLMs easy")
```

How to load this ondisk vector storage?

```python
new_vector_store = FAISS.load_local(
    "today's class faiss index", 
    embeddings=embeddings,
    allow_dangerous_deserialization=True
)

# We need to allow_dangerous_deserialization to load the vector data
```

```python
new_vector_store.similarity_search("langchain")
```

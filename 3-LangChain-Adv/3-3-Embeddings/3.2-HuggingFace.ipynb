#### Sentence Transformers on Hugging Face

- Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. 
- One of the embedding models is used in the HuggingFaceEmbeddings class. 
- We have also added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.

Explore all the embedding models here: https://python.langchain.com/docs/integrations/text_embedding/

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# How do you this model name?
# Just go on the google and search for "sentence transformers embeddings huggingface langchain" you will get this link: https://docs.langchain.com/oss/python/integrations/text_embedding/sentence_transformers
```

```python
embeddings
```

```python
text = "This is a text documents"
query_results = embeddings.embed_query(text)
query_results
```

```python
len(query_results)
```

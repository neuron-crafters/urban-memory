```python
import os
from dotenv import load_dotenv
load_dotenv() 
#https://python.langchain.com/docs/integrations/text_embedding/
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")

embeddings
```

```python
text = "This is a tutorial on OpenAI Embeddings"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
text = "This is a tutorial on OpenAI Embeddings. My name is Agam"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
# Any long the sentence maybe OpenAI Embedding will convert it into size of 3072 vectors.
# And the reason you will understand after learning NLP.
```

```python
# We also define the length of the vector
# The reason is if we want to compare the embedding of two different models at that time we use dimensions
# So incases if a model supports maximum dimensions upto 1024 then we can set 

embeddings = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
text="This is a tutorial on OPENAI embedding"
query_result=embeddings.embed_query(text)
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```
